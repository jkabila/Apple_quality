**Hypothesis test**

```python
sample=df.Size.sample(frac=0.05)
sample
```

```python
t_test,p_value=stats.ttest_1samp(sample,df.Size.mean())
```

```python
if (p_value<0.5):
  print("H0 is rejected")
else:
  print("H0 is accepted")
```

```python
#conti vs conti-2 sample t-test
#size and weight
for i in range(10):
  sample1=df.Weight.sample(frac=0.05)
  sample2=df.Size.sample(frac=0.04)
  t_test,p_value=stats.ttest_ind(sample1,sample2)
  print(p_value,t_test)
  if (p_value<0.05):
    print("H0 is rejected")
  else:
    print("H0 is accepted")
```

```python
#size and Sweetness
for i in range(10):
  sample1=df.Sweetness.sample(frac=0.05)
  sample2=df.Size.sample(frac=0.04)
  t_test,p_value=stats.ttest_ind(sample1,sample2)
  print(p_value,t_test)
  if (p_value<0.05):
    print("H0 is rejected")
  else:
    print("H0 is accepted")
```

**2 sample t-test**

```python
#cont vs conti
for i in range(20):
  sampleA=df.Size.sample(frac=0.04)
  sampleB=df.Weight.sample(frac=0.03)
  t_test, p_value = stats.ttest_ind(sampleA, sampleB)
  if (p_value<0.5):
    print("H0 is rejected")
  else:
    print("H0 is accepted")
```

```python
#crunchiness and sweetness
for i in range(20):
  sampleA=df.Crunchiness.sample(frac=0.04)
  sampleB=df.Sweetness.sample(frac=0.03)
  t_test, p_value = stats.ttest_ind(sampleA, sampleB)
  if (p_value<0.5):
    print("H0 is rejected")
  else:
    print("H0 is accepted")
```

```python
#Ripeness and Juiciness
for i in range(20):
  sampleA=df.Juiciness.sample(frac=0.04)
  sampleB=df.Ripeness.sample(frac=0.05)
  t_test, p_value = stats.ttest_ind(sampleA, sampleB)
  if (p_value<0.5):
    print("H0 is rejected")
  else:
    print("H0 is accepted")
```
